```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

sequence = "I've been waiting for a HuggingFace course my whole life."

model_inputs = tokenizer(sequence)
```

Tokenize a single sequence:

```python
sequence = "I've been waiting for a HuggingFace course my whole life."

model_inputs = tokenizer(sequence)
```

Tokenize Multiple sequences at a time:

```python
sequences = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

model_inputs = tokenizer(sequences)
```

Lets do the padding according to several objectives:

```python
# lets pad the sequences up to the maximum sequence length
model_input = tokenizer(sequences, padding = "longest")

print(model_inputs)
```

```python
# Lets pad the sequences up to the model max length
# (512 for BERT or DistilBERT)

model_input = tokenizer(sequences, padding = "max_length")

print(model_inputs)
```

```python
# Lets pad the sequences up to the specified max length

model_input = tokenizer(sequences, padding = "max_length", max_length = 8)

print(model_inputs)
```

Lets Truncate sequences

```python
# Lets truncate the sequences that are longer than the model max length
# (512 for BERT or DistilBERT)

model_inputs = tokenizer(sequences, truncation = True)

print(model_inputs)
```

```python
# Lets truncate the sequences that are longer than the specified max length

model_inputs = tokenizer(sequences, truncation = True, max_length = 8)

print(model_inputs)
```

```python
## Return PyTorch tensors

model_inputs = tokenizer(sequences, padding = True, return_tensors = "pt")

print(model_inputs)
```

```python
## Return NumPy arrays

model_inputs = tokenizer(sequences, padding = True, return_tensors = "np")

print(model_inputs)
```

Special tokens

```python
sequence1 = "I've been waiting for a HuggingFace course my whole life."
model_inputs = tokenizer(sequence1)
print(model_inputs["input_ids"])
```

```python
tokens = tokenizer.tokenize(sequence1)
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
```

```python
print(tokenizer.decode(model_inputs["input_ids"]))
print(tokenizer.decode(ids))
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
```

```python
sequences = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

tokens = tokenizer(sequences, padding = True, truncation = True, return_tensors = "pt")
```

```python
output = model(**tokens)
```

```python
print(output)
```
