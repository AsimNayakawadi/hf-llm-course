Loading and saving tokenizers

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
tokenizer("Using a Transformer network is simple")
```

```python
tokenizer.save_pretrained("directory_on_my_computer")
```

From words to tokens 

```python
sequence = " Using a Transformer network is simple"
tokens = tokenizer.tokenize(sequence)

print(tokens)
```

From tokens to input IDs

```python
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
```

Decoding

vocabulary indices i.e. Input IDs --> strings or sentences or words

```python
decoded_string = tokenizer.decode([7993, 170, 13809, 23763, 2443, 1110, 3014])
print(decoded_string)
```

The decode method not only converts the indices back to tokens, but also groups together the tokens that were part of the same words to produce a readable sentence.
